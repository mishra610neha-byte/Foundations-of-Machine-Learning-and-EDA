Question 1:  What is the difference between AI, ML, DL, and Data Science? Provide a
A brief explanation of each.

 **ANS**:        
# AI (Artificial Intelligence)

Think of AI as the enterprise-wide strategic umbrella. It’s the overarching capability that enables systems to execute tasks that typically require human intelligence — decision-making, reasoning, problem-solving. AI drives the high-level automation agenda.

# ML (Machine Learning)

Machine Learning is a key operational subset of AI. It leverages data-driven models that learn patterns automatically, enabling systems to improve performance KPIs over time without manual rule-setting. Consider ML as the engine powering predictive capabilities.

# DL (Deep Learning)

Deep Learning is a specialized, high-throughput vertical within ML. It uses neural network architectures to deliver advanced pattern recognition — particularly impactful for unstructured data streams like images, audio, and text. Think of DL as ML on enterprise steroids.
# Data Science

Data Science is the analytics function that operationalizes insights end-to-end: data cleaning, analysis, visualization, modeling, and business storytelling. It aligns data assets with decision-making workflows to unlock measurable business value.

Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent
them?

**ANS**:
# Overfitting
The model learns too much, including noise.       
**Outcome**: Great on training data, poor on new data.        
**Fix**: Use regularization, more data, or simplify the model.

# Underfitting
The model learns too little.       
**Outcome**: Bad on both training and test data         
**Fix**: Use a more complex model or train longer.

# How to detect


**Overfitting**: Training accuracy high, test accuracy low.


**Underfitting**: Both training and test accuracy low.

Question 5: Why is feature scaling important in ML? Compare Min-Max scaling and
Standardization.

**ANS**:
# Feature Scaling Is Important

Feature scaling ensures all numerical features operate on the same playing field, so no single feature overpowers others.
This directly improves model accuracy, training speed, and overall algorithm efficiency.

# Min-Max Scaling (Normalization)

Converts values into a 0 to 1 range.

Very simple, but gets affected by outliers.

Useful for models that rely on distance (KNN, K-Means) or neural networks.

# Standardization (Z-Score Scaling)

Converts data so it has mean = 0 and standard deviation = 1.

Less sensitive to outliers.

Works well for regression, SVM, and PCA pipelines.

Question 6:  Compare Label Encoding and One-Hot Encoding. When would you prefer
one over the other?

**ANS**:
# Label Encoding

Converts each category into a numerical ID (e.g., Red → 0, Blue → 1, Green → 2).

Operational Advantage: Low memory footprint and faster model pipelines.

Risk: Imposes a false ranking between categories.

# One-Hot Encoding

Creates a binary feature vector for every category (e.g., Red → [1,0,0]).

Operational Advantage: Removes any unintended ordinal relationships.

Risk: Expands dimensionality, increasing compute overhead.

# When to Deploy Label Encoding

When the categorical feature has a natural hierarchy (e.g., low < medium < high).

When optimizing for runtime efficiency or working with tree-based models that handle integers gracefully.

# When to Deploy One-Hot Encoding

When the categorical values are purely nominal with no ranking (e.g., city, color).

When you need model-friendly, unbiased categorical representation for linear or distance-based models.
